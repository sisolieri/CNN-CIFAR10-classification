## **Proyecto 1- Añadir profundidad y complejidad a la red (Accuracy: 74,8%)**

Objetivo: subir la accuracy para llegar al 80% minimo requerido en el dataset de test. Al principio no me preocuparé de eventual overfitting, ya que lo que quiero ahora es comprobar que la red sea suficientemente compleja para fitear correctamente los datos que tenemos. Basicamente primero me enfoco en no hacer underfitting.

Implementaciones:
el enfoque del esperimento consite en aumentar la capacidad de extraccion de features, hasta llegar a la extraccion de texturas. He seguido la misma logica que presenta la red VGG-16, una de las mejores redes para clasificación de imagenes, sobretodo por relativa semplicidad en la arquitectura.
Siguen los detalles:

- 3 bloques de convoluciones y al termine de cada bloque un maxpooling para reducir los parametros y quedarme con la informacion mas relevante. Los primeros dos bloques que deberian extraer caracteristicas más genericas, son bloques de 2 convoluciones; el tercer bloque presenta 3 convoluciones porquè he notado un pattern en la arquitectura de VGG-16 que muestra  3 convoluciones para los bloques que extraen features de mas detalle, como las texturas.

 - No puse mas de 3 bloques de convoluciones para evitar de quedarme con demasiados pocos pixeles. De toda forma deberia ser suficiente para detectare texturas, utiles para mejorar la clasificacion.

 - Los filtros en las convoluciones van de menor a mayor y se duplican de un bloque al otro y son multiples de 8 en analogia con el patron de la estructura de VGG. Mas detalle =  mas filtros para extraer mas detalles.

- Aumentado el numero de neuronas en la capa densa de clasificacion para que la parte de la red que se encarga de la clasificacion no se quede atrás respecto a la parte e extraccion de features.

---


**Resultados:**

- La accuracy de train ha subido muchisimo (desde 0.657 a 0.993), gracias a la mayor complejidad de la red.

En el proximo proyecto reduciré el learnig rate para que los pesos puedan cambiar más lentamente. Me espero que el entrenamiento tarde más, pero deberia ser capaz de convergir a un valor estable de error y accuracy.

## 2. Arquitectura de red del modelo


```python
model = ks.Sequential()

model.add(ks.layers.Conv2D(32, (3, 3), strides=1, activation='relu',
                           padding='same', input_shape=(32,32,3)))
model.add(ks.layers.Conv2D(32, (3, 3), strides=1, activation='relu',padding='same'))
model.add(ks.layers.MaxPooling2D((2, 2)))

model.add(ks.layers.Conv2D(64, (3, 3), strides=1, activation='relu',padding='same'))
model.add(ks.layers.Conv2D(64, (3, 3), strides=1, activation='relu',padding='same'))
model.add(ks.layers.MaxPooling2D((2, 2)))

model.add(ks.layers.Conv2D(128, (3, 3), strides=1, activation='relu',padding='same'))
model.add(ks.layers.Conv2D(128, (3, 3), strides=1, activation='relu',padding='same'))
model.add(ks.layers.Conv2D(128, (3, 3), strides=1, activation='relu',padding='same'))
model.add(ks.layers.MaxPooling2D((2, 2)))

model.add(ks.layers.Flatten())
model.add(ks.layers.Dense(256, activation='relu'))
model.add(ks.layers.Dense(10, activation='softmax'))
```

Para revisar un modelo, nos basta con llamar al método **.summary()** del modelo

```python
model.summary()
```

## 3. Optimizador, función error


```python
model.compile(optimizer='Adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

## 4. Preparamos los datos

Cargamos los datos de CIFAR10 de los datasets directamente de las librerias de Keras. Estos ya estan dispuestos en train and test

**Detalle importante:**
> La red neuronal requiere que los inputs sean números reales, y lo haremos forzando la division de los valores de dentro de las matrices 32x32 (que tienen valoress del 0 al 255) por 255.0 (un real)

```python
cifar10 = ks.datasets.cifar10

(x_train, y_train), (x_test, y_test) = cifar10.load_data()

x_train_scaled, x_test_scaled = x_train / 255.0, x_test / 255.0
```

```python
print("Etiquetas de train:", y_train)


# Convertir los datos en arrays de una dimension (vectores) es una buena mejora en el dataset
y_train = y_train.ravel()
y_test = y_test.ravel()


print("Etiquetas de train (en vector):", y_train)
```

Obtenemos un array con todas las labels de CIFAR10

```python
cifar10_labels = [
'airplane', # id 0
'automobile',
'bird',
'cat',
'deer',
'dog',
'frog',
'horse',
'ship',
'truck'
]

print('Number of labels: %s' % len(cifar10_labels))
```

Como vamos a querer ir haciendo validación a la vez que entrenamos (muy practico)

```python
x_val_scaled = x_train_scaled[-10000:]
y_val = y_train[-10000:]

x_train_scaled = x_train_scaled[:-10000]
y_train = y_train[:-10000]
```

Cuando añadimos convoluciones, necesitamos "pre-tratar los datos", porque **la convolución espera una matriz de 4 campos** (más parecido a "imagenes"), en el caso de MNIST. CIFAR10 ya tiene el shape adecuado, así que no tenemos que hacer nada, pero como siempre decimos: es mejor que comprobeis su tamaño.

Por eso, al salir de la Convolution, hay que hacer un Flatten, porque las capas FullDense esperan arrays, no matrices!!

Luego lo imprimimos para ver que todo está correcto

```python
# Validamos el resultado
print('Train: X=%s, y=%s' % (x_train_scaled.shape, y_train.shape))
print('Test: X=%s, y=%s' % (x_test_scaled.shape, y_test.shape))
print('Validation: X=%s, y=%s' % (x_val_scaled.shape, y_val.shape))
```

## 5. Entrenamiento

Ya podemos ponernos a entrenar el modelo!! Ojo, que hemos de entrenar contra los datos convertidos al formato que espera la Convolution.

Empezaremos con 30 epocs, es decir, 30 pasadas completas del dataset (que a su vez sera con mini-batches internamente), fijando un batch de 512

El metodo .fit() nos permite, definir además si disponemos de varias CPUs, GPUs, y si queremos ir validando datos a cada fin de epoch.

```python
t = time.perf_counter()
```

```python
# # EarlyStopping
# earlystop_loss = EarlyStopping(monitor='val_loss', patience=15)
# earlystop_acc = EarlyStopping(monitor='val_accuracy', patience=15)
earlystop_loss = EarlyStopping(monitor='loss', patience=15)

# ModelCheckpoint
checkpoint_path = '/content/drive/MyDrive/Master_Data_Science/Entregables/Entregable_04-Deep_Learning/Models/Model_Checkpoints/Best_accuracy_proyecto_01'
modelcheckpoint_best_acc = ModelCheckpoint(checkpoint_path, monitor='val_accuracy', save_best_only=True)
```

```python
history = model.fit(x_train_scaled, y_train,
                    epochs=100,
                    batch_size= 512,
                    validation_data=(x_val_scaled, y_val),
                    callbacks=[modelcheckpoint_best_acc, earlystop_loss])
```

```python
elapsed_time = datetime.timedelta(seconds=(time.perf_counter() - t))

print('Tiempo de entrenamiento:', elapsed_time)
```

Veamos que tipo de predicciones estoy obteniendo sobre el conjunto de test (vamos a pintar las imagenes y sus clasificaciones)

```python
predictions = model.predict(x_test)
```

Una de las ventajas de Python es que hay montones de funciones y código realizado por terceras personas. Aquí me he fusilado unas bonitas funciones (que he adaptado un poco a mis necesidades) para poder pintar las imagenes, su label (ground truth) y las clasificaciones que hemos realizado.

```python
def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array, true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(predicted_label,
                                100*np.max(predictions_array),
                                true_label),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  predictions_array, true_label = predictions_array, true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')
```

Dibujamos las primeras imagenes, con las predicciones y sus valores reales (un total de 20 imagenes, para no abusar de vuestros laptops)

Coloreamos las prediciones correctas en azul y los fallos en rojo. Aunque primero he impreso las etiquetas para que tengamos una referencia al grafico.

```python
i = 0
for l in cifar10_labels:
    print(i, l)
    i += 1

num_rows = 5
num_cols = 4
start = 650
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
  plt.subplot(num_rows, 2*num_cols, 2*i+1)
  plot_image(i+start, predictions[i+start], y_test, x_test)
  plt.subplot(num_rows, 2*num_cols, 2*i+2)
  plot_value_array(i+start, predictions[i+start], y_test)
plt.tight_layout()
plt.show()
```
